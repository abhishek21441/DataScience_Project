```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Abhishek\\IIITD\Academics\\CSE\\CSE558_DSc\\Project\\Final_Dataset2.csv')

# column_name = 'Week'  # Change this to your actual column name

# # Get and print unique values
# unique_values = df[column_name].unique()
# unique_values.sort()
# print(unique_values)
# print(len(unique_values))

# Specify the columns you want to calculate the average for
columns_to_average = ['NO2', 'SO2', 'CO', 'O3']

# Group by the 'weeks' column and calculate the mean for each group
avg_by_week = df.groupby('Week')[columns_to_average].mean().reset_index()

# Display the resulting DataFrame with average values
print(avg_by_week)
```

```python
Y_NO2 = np.array(avg_by_week['NO2'])
Y_SO2 = np.array(avg_by_week['SO2'])
Y_CO = np.array(avg_by_week['CO'])
Y_O3 = np.array(avg_by_week['O3'])
X = np.array(avg_by_week['Week'])

# Plots
plt.plot(X, Y_NO2, marker='o', linestyle='-')
plt.title('Average NO2 Concentration by Week')
plt.xlabel('Week')
plt.ylabel('Average NO2 Concentration')
plt.grid(True)
plt.xticks(X[::2])
plt.show()

plt.plot(X, Y_SO2, marker='o', linestyle='-')
plt.title('Average SO2 Concentration by Week')
plt.xlabel('Week')
plt.ylabel('Average SO2 Concentration')
plt.grid(True)
plt.xticks(X[::2])
plt.show()

plt.plot(X, Y_CO, marker='o', linestyle='-')
plt.title('Average CO Concentration by Week')
plt.xlabel('Week')
plt.ylabel('Average CO Concentration')
plt.grid(True)
plt.xticks(X[::2])
plt.show()

plt.plot(X, Y_O3, marker='o', linestyle='-')
plt.title('Average O3 Concentration by Week')
plt.xlabel('Week')
plt.ylabel('Average O3 Concentration')
plt.grid(True)
plt.xticks(X[::2])
plt.show()
```

Make it clearer in one graph

```python
# Calculate scaling factor
scaling_factor = np.max(Y_SO2) / np.max(Y_CO)

# Scale list1
scaled_Y_CO = np.array(Y_CO) * scaling_factor

plt.plot(X, Y_NO2, marker='o', linestyle='-',label='NO2', color='blue')
plt.plot(X, Y_SO2, marker='o', linestyle='-',label='SO2', color='red')
plt.plot(X, scaled_Y_CO, marker='o', linestyle='-',label='CO', color='green')
plt.plot(X, Y_O3, marker='o', linestyle='-',label='O3', color='magenta')

plt.title('Average Pollutants Concentration by Week')
plt.xlabel('Week')
plt.ylabel('Average Pollutants Concentration')
plt.grid(True)
plt.xticks(X[::2])
plt.legend()
plt.show()
```
